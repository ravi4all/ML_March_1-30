```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
dataset = pd.read_csv('data/tv_show_viewers_datasest.csv')
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,1:-1]
y = dataset.iloc[:,-1]
```

```python
X.ndim
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.score(X,y)
```

```python
reg.predict([[3,5,2,4,4,3,7,9]])
```

```python
y_pred = reg.predict(X)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
import numpy as np
np.sqrt(mean_squared_error(y,y_pred))
```
